#### Implicit conversion
Implicit class & extension function
1. create a case class student(sno,name,score)
2. process a list of students : studentlist %>70 , studentList %<60.

```scala
case class Student(sno:Int,name:String,score:Int)

import scala.collection.mutable.ListBuffer
val studentList = ListBuffer[Student](
    Student(1, "Alice", 85),
    Student(2, "Bob", 55),
    Student(3, "Charlie", 75),
    Student(4, "David", 45),
    Student(5, "Eve", 90)
)



implicit class StudentOps(studentList:ListBuffer[Student]){
    def %>(threshold:Integer):ListBuffer[Student] = {
        studentList.filter(_.score > threshold)
    }

    def %<(threshold:Integer):ListBuffer[Student] = {
        studentList.filter(_.score < threshold)
    }

    def add(student:(Int,String,Int)) = {

        studentList += Student(student._1,student._2,student._3)
    }

    def add(student:Student) = {
        studentList += student
    }
        
}

// implicit operator overloading
val above70 = studentList %> 70
val below60 = studentList %< 60

// implicit
studentList.add((6,"Aniket",78))
studentList.add(new Student(9,"ANurag",89))
```

```scala
implicit class StudentManager(students:Array[Student]){
    def filterRecords(x:Student => Boolean):Array[Student] = {
      val filteredArray = scala.collection.mutable.ArrayBuffer[Student]()

      for (student <- students) {
        if (x(student)) { 
          filteredArray += student 
        }
      }

      // Converting ArrayBuffer back to Array
      filteredArray.toArray
    }

    def filterStudents(x:Student => Boolean): List[Student] = {
      val filteredList = ListBuffer[Student]()
      for (student <- students) {
        if (x(student)) { 
          filteredList += student 
        }
      }

      filteredList.toList
    }
}

val students = Array(
    Student(1, "Alice", 85),
    Student(2, "Bob", 55),
    Student(3, "Charlie", 75),
    Student(4, "David", 45),
    Student(5, "Eve", 90)
)

students.filterRecords(student => student.score >= 75)
students.filterRecords(student => student.name == "Bob")
```
